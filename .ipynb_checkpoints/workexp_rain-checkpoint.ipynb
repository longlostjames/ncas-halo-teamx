```python
import numpy as np
import os
rain_path = '/gws/pw/j07/ncas_obs_vol2/cao/processing/ncas-rain-gauge-1/data/long-term/2022/202207'

rain_file = os.path.join(rain_path,'ncas-rain-gauge-1_cao_20220719_precipitation_v1.0.nc')
```

```python
import netCDF4 as nc4

DS = nc4.Dataset(rain_file);

import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import xarray as xr

# Open the NetCDF file
ds = xr.open_dataset(rain_file)

# Option 1: Convert entire dataset to DataFrame
df = ds.to_dataframe().reset_index()

# Ensure 'time' column is datetime
df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time');

print(df)
```

```python
fig,ax = plt.subplots(1,1,figsize=(10, 2.5))
ax.plot(df.index,df['thickness_of_rainfall_amount']*6*60)
ax.set_ylabel('Rainfall rate (mm/hr)')
ax.set_xlabel('Time (UTC)')
ax.grid(True)
ax.set_title('2022-07-19 Data from Chilbolton rapid-response rain gauge (unit 1)')
plt.savefig('rainfall_rate_20220719.png',dpi=150);
```
